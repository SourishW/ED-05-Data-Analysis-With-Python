```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("income_data.csv")
```

```python
df.head()
```

```python
df.describe(include='all')
```

```python
df.drop(['fnlwgt', 'native.country'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.replace("?", np.nan, inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
for category in ["workclass", "occupation"]:
    mode = str(df[category].mode()[0])
    df[category].replace(np.nan, mode, inplace=True)
#     print("did it")
    print(mode)
```

```python
df.info()
```

```python
df.head()
```

```python
sns.boxplot(x="income", y="age", data=df)
```

```python
dummies_data = pd.get_dummies(df[["workclass", "education", "marital.status", "occupation", "relationship", "race", "sex"]])
df = pd.concat([dummies_data, df], axis=1)
```

```python
plt.figure(figsize=(50, 30))
sns.heatmap(df.corr(), annot=True, vmin=-1, cmap='RdBu').set_title('Correlation Heatmap')
```
